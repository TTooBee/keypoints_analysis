##### 브레인스토밍
- 현재 발리는 오히려 발리 테이크백을 감지함
- 발리/스트록 구분을 할 필요가 있음
- 백발리는 azimuth값의 변화가 어느정도 눈에 보인다. 포발리는 azimuth값의 변화 거의 없음. 오히려 백발리 테이크백이 더 azimuth값 변화가 큼
- 발리 잡아내려면 공 감지 필수적으로 해야하나? 만약 공 감지를 사용한다면
    - 공 x좌표, y좌표, 사이즈 이렇게 세 파라미터정도 보고, 이 값들의 미분불가능한 점 찾아서 임팩트 또는 바운드 찾아내면 될 듯
- 빠른 공 발리는 손 움직임이 거의 없고, 누르는 발리는 손 움직임이 큼. 즉 발리는 손 움직임이 일관되지 않으므로, 단순히 손 움직임만으로는 샷을 찾아내기가 어려움

- 공 학습 완료까지 대략 70시간 남았음. 2월 16일 오후 3시정도에 끝날 듯 함. 그때까지 뭘 할 수 있을까?
1. 공 감지가 잘 된다는 가정 하에 공 위치를 이용한 분류 알고리즘 구상
    - 공 감지가 잘 된다면, 이게 베스트. 그런데 다 만들어놨는데 감지 엉망이면?
2. 공 감지와 상관없는 부분 만들기
    - 이런 부분이 뭐가 있을까? 
    - shot/non-shot 구분
    - 백코트, 테이크백 나누기
3. 공 감지를 사용하지 않는다는 가정 하에 알고리즘 구상
    - 지금 하고있는 것. 아마도 이거는 힘들듯
    - 그래도 이게 현 상황에서는 가장 안정적인 방법일 듯 하다

- 가장 먼저 할 수 있는 것은(해야 하는 것은) 스트록, 논스트록 구분. 
- 현재는 발리에 스트록 알고리즘 적용해도 샷으로 인정된다. 이 둘을 구분해내야한다.
- 스트록의 특징은?
    - 포핸드의 경우 azimuth의 급격한 증가, 백핸드는 azimuth의 급격한 감소
    - elevation의 증가? elevation은 스타일별로 조금씩 다른듯 하다. 단, 스트록 구간에서는 elevation이 상승하는 구간이 포함됨(맞나??)
    - 스윙스피드 . 얘는 스트록 스윙을 천천히 해버리면 스트록을 못찾아낸다.

- playing/non-playing 분류에 공의 움직임을 포함시키기?
