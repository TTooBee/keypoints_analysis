# 포핸드 백핸드 분류 알고리즘 과정
### 조건 및 현 상황
- 포백 분류는 오른손을 기준으로 한다.
- 아직 오른손의 azimuth 값을 미분 가능하게(부드럽게) 뽑아내지 못하는 경우가 가끔 있다. calibration이 해결되면 좋겠지만, 만약 calibration을 사용하지 못하는 상황이라면 이 부분에 대한 추가적인 알고리즘이 필요할 것이다.
### 알고리즘
1. 오른손의 좌표값 얻는다.
2. 구면좌표계로 변환 후, 지면과 수평한 방향의 각도인 azimuth 값(프레임 수x1의 벡터)을 얻는다.
3. azimuth의 미분을 구하고, 미분이 양수로 유지되는 구간과 음수로 유지되는 구간을 탐색한다. *탐색 결과는 구간의 시작, 끝 인덱스로 저장
4. 탐색한 모든 구간에 대해, 시작 인덱스부터 끝 인덱스까지의 회전각이 0.7pi 이상인 구간만 추출한다.
5. 추출한 인덱스들에 대해, azimuth가 증가하는 구간은 0, 감소하는 구간은 1로 배열 원소를 추가한다.
6. 모든 구간의 기울기를 구한다.
7. 만약 구간이 인접해있다면, 인접한 구간 중 기울기가 가장 큰 값을 가지는 구간을 샷으로 판단한다.
8. 샷으로 판단된 모든 구간을 저장한다.
