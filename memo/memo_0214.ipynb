# 0214 메모
- 우선 사람(사용자)이 카메라 앵글 안에 있을 때, 없을 때를 알아야 함. 
##### lpc를 사용하자
- 사용자가 앵글 밖으로 나간 경우, alphpose에서는 원점(골반)의 좌표에 큰 미분 불가능점이 생긴다. 이를 lpc로 잡아내고, residual을 구해서 residual값이 굉장히 작은(거의 0)인 부분만 사용한다.
- 이때, 사용자가 앵글 밖으로 나가있는 시간이 꽤 길다면 다른 사람에 대해 residual이 0인 구간이 생긴다. 

### 사용자 앵글 밖에 있는 것 + alphapose에서의 오류 감지
- 사용자가 앵글 밖으로 나갈 때 와 들어올 때 골반 keypoint의 좌표가 미분불가능해진다
- 미분 불가능한 점은 lpc를 통해서 알아낼 수 있다
- residual의 값이 큰 부분은 전부 삭제한다
    - 어차피 정규화된 손의 좌표를 사용하기 때문에, 짧은 구간(최대 다섯개 프레임 정도?)의 오류(사람을 잘못 인식한 것)는 신경쓰지 않아도 될 것이다

1. 앵글을 벗어나서 미분 불가능 점이 생겼을 때
- 앵글에서 벗어났다가 다시 들어오면, 다시 돌아왔을 때의 x좌표의 위치는 나갔을 때의 x좌표와 거의 동일해야 한다
    - 예를 들어, x좌표 0 바깥으로 나갔다고 했을 때, 앵글 안으로 다시 들어오려면 0에서 등장해야 할 것이다
    - 이를 이용해서, 만약 앵글 밖(x좌표 0 바깥)으로 원점이 나갔다면, 다시 0에서 등장한 놈을 찾자. 이 부분이 핵심.
2. alphapose의 오류로 미분 불가능 점이 생겼을 때
- 이 구간은 짧다. 그냥 싹 다 날려버리자. 오차는 어쩔 수 없다.
